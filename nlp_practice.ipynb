```python
import re
pattren="[a-zA-Z]*,"
matches=re.findall(pattren,data)
print(matches)
```

```python
doc=nlp(text)
emails=[]
numbers=[]
God=[]
for token in doc:
    if token.like_email:
        emails.append(token.text)
    if token.like_num :
        numbers.append(token.text)
    if token.text=="Joseph" or token.text=="God":
        God.append(token.text)
print(emails)
print(numbers)
print(God)
```
