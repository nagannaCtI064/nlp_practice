```python
import pandas as pd
```

```python
df=pd.read_csv("Ecommerce_data.csv")
```

```python
df.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
v=TfidfVectorizer()
```

```python
def preprocess(text):
    doc=nlp(text)
    filtered_token=[]
    for token in doc:
        if token.is_stop or token.is_punct:
            continue
        filtered_token.append(token.lemma_)
    return " ".join(filtered_token)
```

```python
df["label"].value_counts()
```

```python
df["label_num"]=df["label"].map({
    "Household":0,
    "Electronics":1,
    "Clothing & Accessories":2,
    "Books":3
})
```

```python
df.columns
```

```python
df["preprocess_text"]=df["Text"].apply(preprocess)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report
```

```python
X_train,x_test,y_train,y_test=train_test_split(df["preprocess_text"],df["label_num"],test_size=0.2,random_state=2024,stratify=df["label_num"])
```

```python
y_train.value_counts()
```

```python
y_test.value_counts()
```

```python
clf=Pipeline([
    ('Vectorizertfidf',TfidfVectorizer()),
    ('KNN',KNeighborsClassifier())
])
clf.fit(X_train,y_train)
```

```python
y_pred=clf.predict(x_test)
print(classification_report(y_test,y_pred))
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(cm,annot=True)
plt.xlabel("Predicted Values")
plt.ylabel("Actual Values")
plt.show()
```

```python
data=pd.read_csv("Emotion_classify_Data.csv")
```

```python
data.head()
```

```python
data.shape
```

```python
data["Emotion"].value_counts()
```

```python
data["Emotion_num"]=data["Emotion"].map({
    "anger":0,
    "joy":1,
    "fear":2
})
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
X_train,x_test,y_train,y_test=train_test_split(data["Comment"],data["Emotion_num"],test_size=0.2,random_state=2022,stratify=data["Emotion_num"])
```

```python
len(X_train)
```

```python
len(y_train)
```

```python
len(y_test)
```

```python
len(x_test)
```

```python
clf1=Pipeline([
    ('Vectorizer',CountVectorizer(ngram_range=(3,3))),
    ('KNN',RandomForestClassifier(n_estimators=20))
])
clf1.fit(X_train,y_train)
```

```python
y_pred=clf1.predict(x_test)
```

```python
print(classification_report(y_test,y_pred))
```

```python
from sklearn.naive_bayes import MultinomialNB
clf2=Pipeline([
    ('Vectorizer',CountVectorizer(ngram_range=(1,2))),
    ('MultinomailNB',MultinomialNB())
])
clf2.fit(X_train,y_train)
```

```python
y_pred=clf2.predict(x_test)
```

```python
print(classification_report(y_test,y_pred))
```
